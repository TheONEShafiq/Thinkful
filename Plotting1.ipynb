```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
plt.plot([0,1,2,3])
plt.show()
```

```python
np.random.seed(1221)

df = pd.DataFrame()

df['rand'] = np.random.rand(100)
df['rand_sq'] = df['rand']**2
df['rand_shift'] = df['rand']+2

df['counts_sq'] = df.index ** 2 
df['counts_sqrt'] = np.sqrt(df.index)
```

```python
plt.plot(df['rand'])
plt.show()
```

```python
plt.plot(df['rand'], color='purple')
plt.ylim([-0.1, 1.1])
plt.ylabel('Values')
plt.title('Random Series')
plt.show()
```

```python
plt.plot(df['rand'], color='purple')
plt.plot(df['rand_shift'], color ='green')
plt.ylim([-0.1, 3.1])
plt.ylabel('Values')
plt.title('Random Series')
plt.show()
```

```python
plt.scatter(x=df['rand'], y=df['rand_sq'], color = 'blue')
plt.show()
```

```python
plt.scatter(x=df['rand'], y=df['rand_sq'], color = 'blue', marker ='x', s=10)
plt.scatter(x=df['rand'], y=df['rand_shift'], color = 'green', marker ='x', s=10)
plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.plot(df['rand'], color = 'purple')
plt.ylabel('Values')
plt.title('Random Series')

plt.subplot(1,2,2)
plt.plot(df['rand_shift'], color = 'green')
plt.ylabel("Values")
plt.title('Random Series')

plt.show()
```

```python
x = np.random.normal(10, 5, 1000)

plt.hist(x)
plt.show()



```

```python
plt.hist(x, bins = 40, color = 'red')
plt.xlabel('Random Values')
plt.title('Default Bin Placement Demo')
plt.show()
```

```python
x = np.random.normal(10, 5, 1000)
y = np.random.normal(10, 5, 10000)

plt.hist(x, normed=True, color = 'blue', bins=np.arange(-10,40), alpha=.5)
plt.hist(y, normed=True, color = 'red', bins=np.arange(-10,40), alpha=.5)
plt.title('Manually Sized Bin Demo')
plt.xlabel('Random Values')
plt.show()
```
